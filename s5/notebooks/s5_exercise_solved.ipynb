## Your task: create regression model predicting median house value (target) based on features (data).
#### Building regression model is not so much different than building classigfication model - steps are the same. First we split data into train-test, select model, fit and evaluate its performance - if it's not the best, go for hyperparameter tuning (like grid or random search) or choose another model. Reapeat until you find the best one. You can use regularization (Lasso, Ridge, Elastic Net), data scaling (StandardScaler, MinmaxScaler), etc. to obtain the best results.

## An example of how to solve a task

```python
# scale data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(california_housing.data)
```

```python
from sklearn.model_selection import train_test_split

# train-test split
X_train, X_test, y_train, y_test = train_test_split(scaled_data, california_housing.target, test_size=0.2, random_state=42)
```

```python
# see shape of data
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
''' Helper function for model evaluation '''
def evaluate(reg, X_test):
    #prediction
    y_pred = reg.predict(X_test)

    #evaluation
    #r2
    r2 = r2_score(y_test, y_pred)
    # RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(r2, rmse)
```

```python
# pretty bad, let's try another regressor
evaluate(lr, X_test)
```

```python
from sklearn.linear_model import Ridge
# Ridge Regression
ridge = Ridge(alpha = 1)
ridge.fit(X_train, y_train)
```

```python
ridge.coef_
```

```python
# pretty bad, let's try another regressor
evaluate(ridge, X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
# ensemble model - Random forest
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
# not bad, improve model with hyperparameter tuning
evaluate(rf, X_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Random Layout search to find best hyperparams
n_estimators = [int(x) for x in np.linspace(start = 50, stop = 500, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
```

```python
# Create the random grid params
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# fit and find best params
rf_random = RandomizedSearchCV(estimator = RandomForestRegressor(), param_distributions = random_grid, n_iter = 50, cv = 5, random_state=42)

rf_random.fit(X_train, y_train)
rf_random.best_params_
```

```python
model_random = rf_random.best_estimator_
evaluate(model_random, X_test)
```

```python
from sklearn.tree import plot_tree
#plot one estimator
%matplotlib inline
plot_tree(model_random.estimators_[0])
```

```python
from sklearn.pipeline import make_pipeline
# final pipeline construction
# Pipeline allows you to sequentially apply a list of transformers to preprocess the data and conclude the sequence with a final predictor
pipeline = make_pipeline(StandardScaler(), model_random)
```

```python
pipeline.fit(X_train, y_train).score(X_test, y_test)
```
