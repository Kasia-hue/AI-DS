```python
fig, ax = plt.subplots()

scatter = ax.scatter(
    data=df,
    x="Longitude",
    y="Latitude",
    s=df["MedHouseVal"]*20,
    c="MedHouseVal",
)

legend1 = ax.legend(*scatter.legend_elements(prop="colors"),
                    loc="upper right", title="MedHouseVal")
ax.add_artist(legend1)

_ = plt.title("Median house value depending of\n their spatial location")
plt.show()
```

```python
# Compute the correlation matrix
import seaborn as sns
corr = df.corr()
sns.set(rc={'figure.figsize':(14,10)})
sns.heatmap(corr, vmin=-1, vmax=1, annot=True, cmap="YlGnBu")
```

#### Median house value and median income are highly correlated, average rooms are highly correlated with average bedrooms and longitude and latitude are negatively correlated
